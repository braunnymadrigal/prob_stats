```python
# --------------------------IMPORTING PACKAGES----------------------------------------------------------------------------

import pandas as pd  # the module pandas is getting the nickname 'pd'
from matplotlib import pyplot as plt  # from x package import y module and call it z
import math  # to do the square root

#-------------------------------------------------------------------------------------------------------------------------

#----------------------LITTLE EXAMPLE TO LEARN HOW TO USE pyplot----------------------------------------------------------

def pyplot_usage_example():
    x = [1, 3, 6]  # this little array is gonna work as our x axis in the chart
    y = [1, 4, 9]  # this one gonna work as our y axis
    z = [9, 6, 0]  # this one is going to be another y axis, x can be relate to many data
    plt.plot(x, y)  # relates the x data with the y data, the first data is gonna be located on the x axis
    plt.plot(x, z)  # relates the x data with the z data
    # plt.plot(x, y, 'o')  # the third argument of the plot function is to give the line an specific look
    plt.title("SAMPLE OF TITLE")  # creates a title for this specific just created pyplot
    plt.xlabel("SAMPLE OF xLABEL")  # creates a label for the x axis
    plt.ylabel("SAMPLE OF yLABEL")  # creates a label for the y axis
    plt.legend(["this is x+y", "this is x+z"])  # creates a legend for each plt.plot(), for each line on the pyplot
    plt.show()  # prints the pyplot
    
#-------------------------------------------------------------------------------------------------------------------------

#----------------------LITTLE EXAMPLE TO LEARN HOW TO USE pandas----------------------------------------------------------

def pandas_usage_example():
    # if we press tab when writing a function this ide will help us autocomplete
    my_sample_data = pd.read_csv('sample_data.csv')  # this loads our csv into a variable of a type defined by pandas
    type(my_sample_data)  # this give us the type of any variable, pretty useful
    print(type(my_sample_data))  # types can be printed
    print()  # printing a blank line
    print(my_sample_data)  # the result of the pd.read_type() can be printed
    print()
    my_column_c = my_sample_data.column_c  # this retrieves the data in the third column, that column is called column_c
    my_spec_value = my_column_c.iloc[1]  # this is going to retrieve the second value of the list my_column_c
    print(my_column_c)
    print(my_spec_value)
    # knowing how to use pandas now we can do pyplotS using csv as databases
    plt.plot(my_sample_data.column_a, my_sample_data.column_b)
    plt.show()

#-------------------------------------------------------------------------------------------------------------------------

#--------------LITTLE EXAMPLE TO LEARN HOW TO USE pandas+pyplot-----------------------------------------------------------

def comp_pop_growth_us_and_china_example():
    my_data = pd.read_csv('countries_year_population.csv')
    my_us = my_data[my_data.country == 'United States']  # this will get a data with only the info of the United States
    my_china = my_data[my_data.country == 'China']  # the same as above but now for China
    plt.plot(my_us.year, (my_us.population / 10**6) )  # im dividing the population by a million to see a prettier chart
    plt.plot(my_china.year, (my_china.population / 10**6) )
    # we are now gonna get the pop growth in a percentage related to the pop in the first year of each country
    plt.plot(my_us.year, (my_us.population / my_us.population.iloc[0] * 100) )  # this will give the pop growth in percent
    plt.plot(my_china.year, (my_china.population / my_china.population.iloc[0] * 100) )
    # making the chart look pretty
    plt.title("Comparison of the Population Growth: USA vs CHINA")  
    plt.xlabel("Years") 
    plt.ylabel("Population")
    plt.legend(["US Millions", "China Millions", "US Percentage", "China Percentage"])
    plt.show()  # prints the pyplot

#-------------------------------------------------------------------------------------------------------------------------

#--------------CHART TO SHOW THE RELATIONSHIP BETWEEN 2 VARIABLES---------------------------------------------------------

def relation_between_year_and_population():
    my_data = pd.read_csv('countries_year_population.csv')  # importing the csv
    my_us = my_data[my_data.country == 'United States']  # obtaining the data of the US
    my_china = my_data[my_data.country == 'China']  # obtaining the data of China
    my_india = my_data[my_data.country == 'India']  # obtaining the data of India
    
    plt.plot(my_us.year, (my_us.population / 10**6))  # x axis: us year, y axis: us population in millions
    plt.plot(my_china.year, (my_china.population / 10**6))  # x axis: china year, y axis: china population in millions
    plt.plot(my_india.year, (my_india.population / 10**6))  # x axis: india year, y axis: india population in millions
    
    plt.title("Relation Between the Year and the Population of the 3 Most Populated Countries")  # adding a title
    plt.xlabel("Years")  # adding a label to the x axis
    plt.ylabel("Population in Millions")  # adding a label to the y axis
    plt.legend(["USA", "CHINA", "INDIA"])  # legends to improve readability
    plt.show()

#-------------------------------------------------------------------------------------------------------------------------

#--------IMPORTING one-person-households-vs-gdp-per-capita.csv INTO A DATAFRAME AND RETURNING IT--------------------------

def get_spec_database():
    my_data = pd.read_csv('one-person-households-vs-gdp-per-capita.csv')  # my_data is the dataframe obtained from the csv
    return my_data  # returning the data frame to modularize, a little, this code

#-------------------------------------------------------------------------------------------------------------------------

#---------PRINT 3 CHOSEN VALUES FROM THE SPEC DATAFRAME-------------------------------------------------------------------

def print_3_chosen_values(my_data):
    # loc: uses the label, iloc: uses an index
    try:
        my_spec_value_1 = my_data.iloc[784, 0]  # get value on column 1, row 786
        my_spec_value_2 = my_data.iloc[776, 5]  # get value on column 6, row 778
        my_spec_value_3 = my_data.iloc[1290, 6]  # get value on column 7, row 1292
    
        print("Column 1, Row 786: ", my_spec_value_1)
        print("Column 6, Row 778: ", my_spec_value_2)
        print("Column 7, Row 1292: ", my_spec_value_3)
    except:
        print("Your dataframe can not be use on this context")

#-------------------------------------------------------------------------------------------------------------------------

#--------------PRINT THE COLUMN NAMES OF THE DATAFRAME PARAMETER----------------------------------------------------------

def print_column_names(my_data):
    try:
        for col in my_data.columns:
            print(col, " ")
    except:
        print("Your dataframe can not be use on this context")

#-------------------------------------------------------------------------------------------------------------------------

#----------------MODIFY SPECIFIC COLUMN NAMES OF THE DATAFRAME PARAMETER--------------------------------------------------

def modify_column_names(my_data):
    try:
        my_data.rename(columns = {'GDP per capita, PPP (constant 2011 international $)':'GDP per capita', 'Share of one person households':'Share 1p household'}, inplace = True)
    except:
        print("Your dataframe can not be use on this context")
    return my_data

#-------------------------------------------------------------------------------------------------------------------------

#-----------------PRINT NUMBER OF ROWS AND COLUMNS------------------------------------------------------------------------

def print_num_rows_columns(my_data):
    try:
        num_rows = my_data.shape[0]  # there are other ways to obtain this value, this one is simplier
        num_columns = my_data.shape[1]  # there are other ways to obtain this value, this one is simplier
        print("Number of Rows: ", num_rows, "Number of Columns: ", num_columns)  
    except:
        print("Your dataframe can not be use on this context")

#-------------------------------------------------------------------------------------------------------------------------

#------------PRINT A SAMPLE OF THE DATAFRAME PARAMETER, THE NUM PARAMETER INDICATES THE SAMPLE SIZE-----------------------

def print_sample(my_data, num):
    try:
        print(my_data.head(num))  # print a sample of the dataframe using the .head() and inside it a value
    except:
        print("Your dataframe can not be use on this context")

#-------------------------------------------------------------------------------------------------------------------------


#-----------ELIMINATE ROWS WITH NULL VALUES-------------------------------------------------------------------------------

def delete_rows__with_null_values(my_data):
    try:
        my_data.dropna(subset=['GDP per capita'], inplace=True)  # remove all rows that contains null values in GDP
        my_data.dropna(subset=['Share 1p household'], inplace=True)  # 'inplace=True' indicates to modify the 'my_data'
    except:
        print("Your dataframe can not be use on this context")    
    return my_data

#-------------------------------------------------------------------------------------------------------------------------

#-----------ELIMINATE ROWS WHICH YEAR IS BELOW 2015-----------------------------------------------------------------------

def delete_rows__below_year_2015(my_data):
    try:
        my_data = my_data[my_data.Year > 2015]  # preserves only the my_data which "Year" rows are >= than 2015
    except:
        print("Your dataframe can not be use on this context")    
    return my_data

#-------------------------------------------------------------------------------------------------------------------------

#-----------ELIMINATE REPEATED COUNTRIES. KEEP FOR EACH COUNTRY IT'S MOST RECENT DATA-------------------------------------

def delete_old_repeated_data(my_data):
    try:
        my_data = my_data.sort_values(['Year'], ascending=[False])  # sort my_data according to column 'Year'
        list_countries = []  # list to check if a country has already been seen
        for row in my_data.itertuples():  # iterates for each row. row is a tuple
            aux_country = row[1]  # assigns the country in row to aux_country that tells current country
            if aux_country in list_countries:  # checks if aux_country is in the list of seen countries
                my_data.drop(row[0], inplace=True)  # drop the row containing the repeated country, row[0] == row index
            else:
                list_countries.append(aux_country)  # if false, the country is added to the list of seen countries
    except:
        print("Your dataframe can not be use on this context")
    return my_data

#-------------------------------------------------------------------------------------------------------------------------

#----------------------PRINT EACH ROW OF THE DATAFRAME PARAMETER----------------------------------------------------------

def print_each_row(my_data):
    try:
        for row in my_data.itertuples():  # iterates for each row. row is a tuple
            print(row)
    except:
        print("Your dataframe can not be use on this context")    

#-------------------------------------------------------------------------------------------------------------------------

#----------------------------------CHART TO SHOW THE RELATION BETWEEN 1PH AND GDP-----------------------------------------

def relation_between_1ph_and_gdp(my_data):
    try:
        my_data = my_data.sort_values(['Share 1p household'])  # sort my_data according to column '1ph'
        plt.plot(my_data['Share 1p household'], my_data['GDP per capita'])  # plot the 2 columns
        # making the chart look pretty
        plt.title("Relation between 'Share of 1p Households' and 'GDP per Capita'")  
        plt.xlabel("share of 1p households") 
        plt.ylabel("gdp per capita")
        plt.legend(["1ph, gdp"])
        plt.show()  # prints the pyplot
    except:
        print("Your dataframe can not be use on this context")    

#-------------------------------------------------------------------------------------------------------------------------

#-----------ELIMINATE ROWS WHICH GDP IS ABOVE 70000-----------------------------------------------------------------------

def delete_gdp_outliers(my_data):
    try:
        my_data = my_data[my_data['GDP per capita'] <= 70000]  # delete rows which gdp > 70000
    except:
        print("Your dataframe can not be use on this context")    
    return my_data

#-------------------------------------------------------------------------------------------------------------------------

#-------------CALCULATE THE CORRELATION (PEARSON) BETWEEN 1PH AND GDP-----------------------------------------------------

def get_correlation(my_data):
    correlation = 0
    try:
        my_data = my_data.sort_values(['Share 1p household'])  # sort my_data according to column '1ph'
        num_rows = my_data.shape[0]  # there are other ways to obtain this value, this one is simplier
        x_prom = 0
        y_prom = 0
        
        for row in my_data.itertuples():  # iterates for each row
            x_prom = x_prom + row[6]
            y_prom = y_prom + row[7]
            
        x_prom = x_prom / num_rows
        y_prom = y_prom / num_rows
        val_sup = 0
        val_inf = 0
        x_inf = 0
        y_inf = 0
        
        for row in my_data.itertuples():  # iterates for each row
            val_sup = val_sup + (row[6] - x_prom) * (row[7] - y_prom)
            x_inf = x_inf + (row[6] - x_prom) * (row[6] - x_prom)
            y_inf = y_inf + (row[7] - y_prom) * (row[7] - y_prom)

        val_inf = math.sqrt(x_inf) * math.sqrt(y_inf)
        correlation = val_sup / val_inf
        #correlation = my_data['Share 1p household'].corr(my_data['GDP per capita'])
    except:
        print("Your dataframe can not be use on this context")            
    return correlation

#-------------------------------------------------------------------------------------------------------------------------

#------------MAIN/INTERFACE------------------------------------------------------------

def print_menu():
    print("frame: this is how this interface calls the dataframe generated from the csv '1ph-vs-gdp'")
    print("0. exit")
    print("1. show chart with the relation between 2 variables (using another csv for this exercise)")
    print("2. (frame) print 3 fixed values")
    print("3. (frame) print column names")
    print("4. (frame) modify column names")
    print("5. (frame) print number of rows and columns. print head(100)")
    print("6. (frame) removes rows with null vals on 1ph or gdp columns. removes rows which year is < 2015")
    print("7. (frame) removes repeated countries, for each country keep the one with the latest year")
    print("8. (frame) show chart with the relation between 1ph and gdp")
    print("9. (frame) print correlation between 1ph and gdp")
    print("10. (frame) (without outliers) show chart with the relation between 1ph and gdp")
    print("11. (frame) (without outliers) print correlation between 1ph and gdp")

def main():
    my_data = get_spec_database()  # charge the database into the my_data dataframe
    while True:
        print_menu()
        option = input("Pick an option: ")

        if option == '0':
            print("BYE")
            break
        elif option == '1':
            relation_between_year_and_population()
            print()
        elif option == '2':
            aux_data = my_data
            print_3_chosen_values(aux_data)
            print()
        elif option == '3':
            aux_data = my_data
            print_column_names(aux_data)
            print()
        elif option == '4':
            aux_data = my_data
            aux_data = modify_column_names(aux_data)
            print_column_names(aux_data)
            my_data = aux_data
            print()
        elif option == '5':
            aux_data = my_data
            aux_data = modify_column_names(aux_data)
            print_num_rows_columns(aux_data)
            print_sample(aux_data, 100)
            print()
        elif option == '6':
            aux_data = my_data
            aux_data = modify_column_names(aux_data)
            aux_data = delete_rows__with_null_values(aux_data)
            aux_data = delete_rows__below_year_2015(aux_data)
            print_sample(aux_data, 100)
            my_data = aux_data
            print()
        elif option == '7':
            aux_data = my_data
            aux_data = modify_column_names(aux_data)
            aux_data = delete_rows__with_null_values(aux_data)
            aux_data = delete_rows__below_year_2015(aux_data)
            aux_data = delete_old_repeated_data(aux_data)
            print_sample(aux_data, 100)
            print_num_rows_columns(aux_data)
            my_data = aux_data
            print()
        elif option == '8':
            aux_data = my_data
            aux_data = modify_column_names(aux_data)
            aux_data = delete_rows__with_null_values(aux_data)
            aux_data = delete_rows__below_year_2015(aux_data)
            aux_data = delete_old_repeated_data(aux_data)
            relation_between_1ph_and_gdp(aux_data)
            print()
        elif option == '9':
            aux_data = my_data
            aux_data = modify_column_names(aux_data)
            aux_data = delete_rows__with_null_values(aux_data)
            aux_data = delete_rows__below_year_2015(aux_data)
            aux_data = delete_old_repeated_data(aux_data)
            correlation = get_correlation(aux_data)
            print(correlation)
            print()
        elif option == '10':
            aux_data = my_data
            aux_data = modify_column_names(aux_data)
            aux_data = delete_rows__with_null_values(aux_data)
            aux_data = delete_rows__below_year_2015(aux_data)
            aux_data = delete_old_repeated_data(aux_data)
            aux_data = delete_gdp_outliers(aux_data)
            relation_between_1ph_and_gdp(aux_data)
            print()
        elif option == '11':
            aux_data = my_data
            aux_data = modify_column_names(aux_data)
            aux_data = delete_rows__with_null_values(aux_data)
            aux_data = delete_rows__below_year_2015(aux_data)
            aux_data = delete_old_repeated_data(aux_data)
            aux_data = delete_gdp_outliers(aux_data)
            correlation = get_correlation(aux_data)
            print(correlation)
            print()
        else:
            print("Invalid option.")
            print()
main()

#--------------------------------------------------------------------------------------
```
